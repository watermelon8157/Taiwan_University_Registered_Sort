```python
#http://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html
#載入套件
from ipywidgets import *#網頁互動
#widgets 元素 #HBox 橫軸 #VBox 縱軸 #Layout 畫布
from IPython.core.display import display, HTML #輸出顯示
from functools import partial #互動程式用 #http://blog.blackwhite.tw/2013/02/python-functoolspartial.html
from DBA import * #資料庫連線設定

from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #將預設字體改用SimHei字體
import matplotlib.pyplot as plt
```

```python
#產生畫面及按鈕功能
cols = ['學年度','學校類別','設立別','學校代碼','學校名稱','日間','學制別','科系代碼','科系名稱','學生總計','新生註冊率','一年級人數','縣市名稱','體系別']
class General_functions(object):
    #####建構子(設定畫面)
    def __init__(self):
        #//判斷是否已經設定資料了
        self.isSetData = False
        self.DBA = None
        #///查詢項目
        self.Label1 = widgets.Label(value="查詢方式:")
        self.dpd1 = widgets.Dropdown(
            options=['使用CSV檔案', '使用MySQL資料庫'],disabled=False,layout=Layout(width='150px')
        )
        #///學年度
        self.Label2 = widgets.Label(value="學年度:")
        self.sYear = widgets.Dropdown(
            options=['請選擇'] ,disabled=False,layout=Layout(width='100px')
        )
        self.Label2_1 = widgets.Label(value="至:")
        self.eYear = widgets.Dropdown(
            options=['請選擇'] ,disabled=False,layout=Layout(width='100px')
        )  
        #///學校類別
        self.Label3 = widgets.Label(value="學校類別:")
        self.schoolChk1 = widgets.Checkbox(
            value=True,description='一般大學',disabled=False
        )
        self.schoolChk2 = widgets.Checkbox(
            value=False,description='技專校院',disabled=False
        )
         #///設立別
        self.Label4 = widgets.Label(value="設立別:")
        self.school_type1 = widgets.Checkbox(
            value=True,description='日間部(學制)',disabled=False
        )
        self.school_type2 = widgets.Checkbox(
            value=False,description='進修部(學制)(含在職專班)',disabled=False
        )
        #///學制別
        self.Label5 = widgets.Label(value="學制別:")
        self.school_class = widgets.Checkbox(
            value=True,description='學士班(含四技)',disabled=True
        )
        self.Label6 = widgets.Label(value="學校名稱:")
        self.school_name = widgets.Dropdown(options=['請選擇'])
        #///系統訊息
        self.Label7 = widgets.Label(value="系統訊息:")
        self.Valid = widgets.Valid(value=True,description='產生查詢介面!',layout=Layout(width='600px'))
        #///產生結果
        self.html = widgets.HTML(value="")
        #///圖表選擇
        self.Label9 = widgets.Label(value="選擇圖表:")
        self.chartChk = widgets.Dropdown(
            options=['長條圖', '圓餅圖'],disabled=False,layout=Layout(width='150px')
        )
        self.chartHtml = widgets.HTML(value="")
    #///錯誤訊息
    def isValid(self, pMsg=''):
        if pMsg != '':
            self.Valid.description = pMsg
        else:
            self.Valid.description = '程式錯誤，請洽資訊人員!'
        self.Valid.value =False
     #///當按鈕按下去的時候要做的事情
    def on_button0_clicked(self, button):
        button.disabled = True
        self.isValid('資料載入中....')
        self.DBA = DBA(self.dpd1.value)
        if self.DBA.getLastError() != '':
            self.isValid(self.DBA.getLastError())
        else:
            self.Valid.description = '資料設定成功!'
            self.Valid.value =True
            self.isSetData = True
            self.setDropDownList()
        button.disabled = False
    def getTableHTML(self, a):
        pStr = '<table border="1">' 
        pStr += '<tr>' 
        for i in cols:
            pStr += '<td>' +str(i)+'</td>'
        pStr += '</tr>'
        for i in a.values:
            pStr += '<tr>' 
            for j in i:
                pStr += '<td>' +str(j)+'</td>' 
            pStr += '</tr>'
        pStr += '</table>'  
        return pStr
    #設定下拉式選單
    def setDropDownList(self):
        Year = list(set([str(int(x)) for x in list(self.DBA.getData()['學年度']) if str(x) != 'nan']))
        Year.append('請選擇');Year.reverse()
        school_name = list(set([str(x) for x in list(self.DBA.getData()['學校名稱']) if str(x) != 'nan']))
        school_name.append('請選擇');school_name.reverse()
        self.sYear.options = Year;self.sYear.value =self.sYear.options[1]
        self.eYear.options = Year;self.eYear.value =self.eYear.options[1]
        self.school_name.options = school_name
    
    #///當按鈕按下去的時候要做的事情
    def on_button1_clicked(self, button):
        if self.isSetData:
            self.html.value = self.getTableHTML(self.DBA.getData().head())
        else:
            self.isValid()
    #///當按鈕按下去的時候要做的事情
    def on_button2_clicked(self, button):
        if self.isSetData:
            plt.plot([1, 2, 3], [4, 5, 6])
            plt.xlabel("中文哈哈哈")
            plt.ylabel("中文這是中文")
            plt.title("我是天才")
            plt.savefig('button2.png', dpi=300)
            self.html.value = '<img src="button2.png" alt="no image!" width="500" height="377">'
        else:
            self.isValid()
    #///將顯示畫面印出
    def draw_board(self):
        button0 = widgets.Button(description='設定資料取得方式')
        button0.on_click(partial(self.on_button0_clicked))
        button1 = widgets.Button(description='查詢報表' )
        button1.on_click(partial(self.on_button1_clicked))
        button2 = widgets.Button(description='產生圖表' )
        button2.on_click(partial(self.on_button2_clicked))
        self.Accordion =  widgets.Accordion(children=[button1,VBox([HBox([self.Label9,self.chartChk,button2]),self.chartHtml])])
        self.Accordion.set_title(0, '產生報表(dataFrame)')
        self.Accordion.set_title(1, '產生圖表(Chart)')
        #將顯示畫面印出......
        HTML(display( VBox([HBox([self.Label7,self.Valid]),
                            HBox([self.Label1,self.dpd1,button0]),
                           HBox([self.Label2,self.sYear,self.Label2_1,self.eYear]),
                           HBox([self.Label3,self.schoolChk1,self.schoolChk2]),
                           HBox([self.Label4,self.school_type1,self.school_type2]),
                           HBox([self.Label5,self.school_class]),
                           HBox([self.Label6,self.school_name]),
                           self.Accordion,
                           self.html])))
```
